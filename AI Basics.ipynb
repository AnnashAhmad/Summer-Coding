Write a function in python that receives a string and returns True if that string is a palindrome and False otherwise.

```python
def isPalindrome(word):
    temp=word[::-1]
    if temp.capitalize()==word.capitalize():
        return True
    else:
        return False
print(isPalindrome("deed"))
```

The derivate of a function ùëì(ùë•)is a measurement of how quickly the function ùëìchanges with respect to change in its domain ùë•. This measurement can be approximated by the following relation,
Where h represents a small increment in x. You have to prove the following relation,
Imagine x being a list that goes from ‚Äìpi to pi with an increment of 0.001. You can approximate the derivative by using the following approximation,
In your case, assume h = 0.001. That is at each point in x, compute the right hand side of above equation and compare whether the output value is equivalent to cos(x). Also print the corresponding values of ùëëùëëùë•sin(ùë•)and cos(x) for every point. Type ‚Äò‚Äôfrom math import *‚Äô‚Äô at the start of your program to use predefined values of pi, and sin and cos functions. What happens if you increase the interval h from 0.001 to 0.01 and then to 0.1?
Moreover, also plot your results using the matplotlib library 

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin, cos, pi

# Define the function for derivative approximation
def derivative_approximation(x, h):
    return (sin(x + h) - sin(x)) / h

# Define the function for the exact derivative
def exact_derivative(x):
    return cos(x)

# Define the range of x values from -pi to pi with an increment of 0.001
x_values = np.arange(-pi, pi, 0.001)

# Initialize lists to store results
approx_derivatives = []
exact_derivatives = []

# Compute derivatives and compare with exact values for each x
for x in x_values:
    approx = derivative_approximation(x, 0.001)
    exact = exact_derivative(x)
    approx_derivatives.append(approx)
    exact_derivatives.append(exact)
    print(f"x = {x:.3f}: Approximate Derivative = {approx:.3f}, Exact Derivative = {exact:.3f}")
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, approx_derivatives, label='Approximate Derivative')
plt.plot(x_values, exact_derivatives, label='Exact Derivative (cos(x))')
plt.title('Comparison of Approximate and Exact Derivatives of sin(x)')
plt.xlabel('x')
plt.ylabel('Derivative')
plt.legend()
plt.grid(True)
plt.show()
```

If you increase the interval h from 0.001 to 0.01 and then to 0.1, the accuracy of the derivative approximation will decrease. This is because a larger value of h leads to a larger step size in the approximation, causing it to deviate more from the actual derivative.
